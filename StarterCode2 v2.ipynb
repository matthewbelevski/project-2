```python
import pandas as pd
import os
import glob
import pandas as pd
```

```python
traffic_df = pd.read_csv("Traffic/combined_csv.csv" )
traffic_df.head(20)
```

```python
#convert negative numbers which are assumed to be error codes to zero in AM Peak Fields
traffic_df['V23'][traffic_df['V23'] < 0] = 0
traffic_df['V24'][traffic_df['V24'] < 0] = 0
traffic_df['V25'][traffic_df['V25'] < 0] = 0
traffic_df['V26'][traffic_df['V26'] < 0] = 0
traffic_df['V27'][traffic_df['V27'] < 0] = 0
traffic_df['V28'][traffic_df['V28'] < 0] = 0
traffic_df['V29'][traffic_df['V29'] < 0] = 0
traffic_df['V30'][traffic_df['V30'] < 0] = 0
traffic_df['V31'][traffic_df['V31'] < 0] = 0
```

```python
#combines the values of the AM Peak time of 6:00am to 8:00am
traffic_df["AM_Peak"] = (traffic_df["V23"] + traffic_df["V24"]  + traffic_df["V25"]  + traffic_df["V26"]  + traffic_df["V27"] 
                        + traffic_df["V28"] + traffic_df["V29"]  + traffic_df["V30"]  + traffic_df["V31"] )
```

```python
#convert negative numbers which are assumed to be error codes to zero in PM Peak Fields
traffic_df['V63'][traffic_df['V63'] < 0] = 0
traffic_df['V64'][traffic_df['V64'] < 0] = 0
traffic_df['V65'][traffic_df['V65'] < 0] = 0
traffic_df['V66'][traffic_df['V66'] < 0] = 0
traffic_df['V67'][traffic_df['V67'] < 0] = 0
traffic_df['V68'][traffic_df['V68'] < 0] = 0
traffic_df['V69'][traffic_df['V69'] < 0] = 0
traffic_df['V70'][traffic_df['V70'] < 0] = 0
traffic_df['V71'][traffic_df['V71'] < 0] = 0
```

```python
#combines the values of the PM Peak time of 4:00pm to 6:00pm
traffic_df["PM_Peak"] = (traffic_df["V63"] + traffic_df["V64"]  + traffic_df["V65"]  + traffic_df["V66"]  + traffic_df["V67"] 
                        + traffic_df["V68"] + traffic_df["V69"]  + traffic_df["V70"]  + traffic_df["V71"] )
```

```python
#select only the columns we want
traffic_df_clean = traffic_df[["NB_SCATS_SITE", "QT_INTERVAL_COUNT", "NB_DETECTOR", "NM_REGION", "CT_RECORDS", "QT_VOLUME_24HOUR", "CT_ALARM_24HOUR", "AM_Peak", "PM_Peak"]]
```

```python
traffic_df_clean.head()
```

```python
#removes 0:00 from the date
traffic_df_clean['QT_INTERVAL_COUNT'] = traffic_df_clean['QT_INTERVAL_COUNT'].str.replace(' 0:00', '')
```

```python
#rename column to Date for later use when merging
traffic_df_clean = traffic_df_clean.rename(columns={"QT_INTERVAL_COUNT" : "Date"})
```

```python
#group by the region, date and scat site
traffic_df_clean_2 = traffic_df_clean.groupby(["NM_REGION", "Date"]).agg({"QT_VOLUME_24HOUR" : ["sum"], "AM_Peak" : ["sum"], "PM_Peak" : ["sum"]})
```

```python
traffic_df_clean_2.head()
```

```python
#rename columns
traffic_df_clean_2.columns = ['QT_VOLUME_24HOUR', 'AM_PEAK_VOLUME', 'PM_PEAK_VOLUME']
#reset the index
traffic_df_clean_2 = traffic_df_clean_2.reset_index()
```

```python
#import weather csv as dataframe
weather_df = pd.read_csv('weatherAUS.csv/weatherAUSJuly2016.csv')
```

```python
weather_df.head(20)
```

```python
#filter to only include Melbourne
weather_df_melb = weather_df[weather_df['Location']=="Melbourne"]
```

```python
#select only the columns we want
weather_df_melb_clean = weather_df_melb[["Date", "Location", "MaxTemp", "Rainfall", "WindGustSpeed", "Humidity9am", "Humidity3pm"]]
```

```python
weather_df_melb_clean.head()
```

```python
#merge weather and traffic dataframes on date
traffic_df_clean_3 = pd.merge(traffic_df_clean_2, weather_df_melb_clean, on = "Date", how = "inner")
```

```python
traffic_df_clean_3
```

```python
#load in nm region data which includes the region name and lat and long
nm_region = pd.read_csv('Traffic_Other/NM_Region.csv', encoding= 'unicode_escape')
```

```python
nm_region
```

```python
#merge nm region data onto combined traffic and weather dataframe
traffic_df_clean_4 = pd.merge(traffic_df_clean_3, nm_region, on = "NM_REGION", how = "inner")

```

```python
#filter out non melbourne regions
traffic_df_clean_5 = traffic_df_clean_4[(traffic_df_clean_4["NM_REGION"] != "BEN") & 
                                       (traffic_df_clean_4["NM_REGION"] != "GEE") &
                                        (traffic_df_clean_4["NM_REGION"] != "GE2") &
                                       (traffic_df_clean_4["NM_REGION"] != "VI2") &
                                       (traffic_df_clean_4["NM_REGION"] != "VIC") &
                                       (traffic_df_clean_4["NM_REGION"] != "WV1") &
                                       (traffic_df_clean_4["NM_REGION"] != "WV2") &
                                       (traffic_df_clean_4["NM_REGION"] != "DUP") &
                                       (traffic_df_clean_4["NM_REGION"] != "DIO")]
```

```python
traffic_df_clean_5.head()
```

```python
#save newly combined dataframe to csv
traffic_df_clean_5.to_csv("CombinedDataTrafficWeatherV3.csv" )
```
